```python
import networkx as nx
# création d'un graph vierge
G = nx.Graph()
# création de noeuds
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from([(4, {'color': 'red'}),(5, {'color': 'green'})])

# afficher les noeuds : font_weight = poids de la police  bold = gras
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
""" bizarre que les couleurs ne soient pas bonnes sur les noeuds 4 et 5 """
```

```python
G.add_edges_from([(3, 4), (4, 5)], color='red')
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_edge(1, 2)  # ajout de edge=bord=arete entre sommets 1 et 2 à G
e = (2, 3)        # tuple de valeurs de bords ou  d'aretes entre sommets 2 et 3 
G.add_edge(*e)    # déballe les bords du tuple*

# affiche sommet et aretes
nx.draw(G, with_labels=True, font_weight='bold')
```
